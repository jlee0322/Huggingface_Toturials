```python
from transformers import BertTokenizer

#加載預訓練字典和分詞方法
tokenizer = BertTokenizer.from_pretrained(
    pretrained_model_name_or_path='bert-base-chinese',
    cache_dir=None,
    force_download=False,
)

sents = [
    '今天應該不會下雨。',
    '兩個禮拜以後，會有一部電影上映。',
    '學校附近有什麼好吃的?',
    '要前往高鐵站，你必須搭那班公車。',
    '老乾媽配飯真是不錯!',
]

tokenizer, sents
```

```python
#編碼兩個句子
out = tokenizer.encode(
    #第一句&第二句
    text=sents[0],
    text_pair=sents[1],  

    #當句子長度大於max_length時，截斷
    truncation=True,

    #一律padding到max_length長度
    padding='max_length',
    add_special_tokens=True,
    max_length=30,
    return_tensors=None,
)
print(out)
print('-'*10 + '我是分隔線' + '-'*10)
tokenizer.decode(out)
```

```python
#編碼函數擴充版(return更多參數)
out = tokenizer.encode_plus(
    text=sents[0],
    text_pair=sents[1],
    
    #當句子長度大於max_length時，截斷
    truncation=True,

    #一律paddibg到max_length長度
    padding='max_length',
    max_length=30,
    add_special_tokens=True,

    #可取值tf,pt,np,預設返回list
    return_tensors=None,

    #return token_type_ids
    return_token_type_ids=True,

    #return attention_mask
    return_attention_mask=True,

    #return special_tokens_mask
    return_special_tokens_mask=True,

    #return length
    return_length=True,
)

for k, v in out.items():
    print(k, ':', v)
#input_ids 就是編碼後的詞
#token_type_ids 第一個句子和特殊符號的位置是0,第二個句子的位置是1
#special_tokens_mask 特殊符號的位置是1,其他位置是0
#attention_mask pad的位置是0,其他位置是1
#length  return句子長度
print('-'*10 + '我是分隔線' + '-'*10)
tokenizer.decode(out['input_ids'])
```

```python
#批次編碼句子
out = tokenizer.batch_encode_plus(
    batch_text_or_text_pairs=[sents[0], sents[1]],
    add_special_tokens=True,

    #當句子長度大於max_length時，截斷
    truncation=True,

    #一律padding到max_length長度
    padding='max_length',
    max_length=15,

    #可取值tf,pt,np,預設返回list
    return_tensors=None,

    #return token_type_ids
    return_token_type_ids=True,

    #return attention_mask
    return_attention_mask=True,

    #return special_tokens_mask
    return_special_tokens_mask=True,

    #return length
    return_length=True,
)
for k, v in out.items():
    print(k, ':', v)
#input_ids 就是編碼後的詞
#token_type_ids 第一個句子和特殊符號的位置是0,第二個句子的位置是1
#special_tokens_mask 特殊符號的位置是1,其他位置是0
#attention_mask pad的位置是0,其他位置是1
#length  return句子長度
print('-'*10 + '我是分隔線' + '-'*10)
tokenizer.decode(out['input_ids'][0]), tokenizer.decode(out['input_ids'][1])
```

```python
#批次編碼成對的句子
out = tokenizer.batch_encode_plus(
    batch_text_or_text_pairs=[(sents[0], sents[1]), (sents[2], sents[3])],
    add_special_tokens=True,

    #當句子長度大於max_length時,截斷
    truncation=True,

    #一律padding到max_length的長度
    padding='max_length',
    max_length=30,

    #可取值tf,pt,np,預設返回list
    return_tensors=None,

    #return token_type_ids
    return_token_type_ids=True,

    #return attention_mask
    return_attention_mask=True,

    #return special_tokens_mask
    return_special_tokens_mask=True,

    #return length
    return_length=True,
)

for k, v in out.items():
    print(k, ':', v)
#input_ids 就是編碼後的詞
#token_type_ids 第一個句子和特殊符號的位置是0,第二個句子的位置是1
#special_tokens_mask 特殊符號的位置是1,其他位置是0
#attention_mask pad的位置是0,其他位置是1
#length  return句子長度
print('-'*10 + '我是分隔線' + '-'*10)
tokenizer.decode(out['input_ids'][0]), tokenizer.decode(out['input_ids'][1])
```

```python
#獲取字典
the_dict = tokenizer.get_vocab()

type(the_dict), len(the_dict), '月光' in the_dict,
```

```python
#添加新詞
tokenizer.add_tokens(new_tokens=['月光', '希望'])

#添加新token
tokenizer.add_special_tokens({'eos_token': '[EOS]'})

the_dict = tokenizer.get_vocab()

type(the_dict), len(the_dict), the_dict['月光'], the_dict['希望'], the_dict['[EOS]']
```

```python
#編碼新添加的詞
out = tokenizer.encode(
    text='月光的新希望[EOS]',
    text_pair=None,

    #當句子長度大於max_length時，截斷
    truncation=True,

    #一律padding到max_length長度
    padding='max_length',
    add_special_tokens=True,
    max_length=10,
    return_tensors=None,
)

print(out)

tokenizer.decode(out)
```
